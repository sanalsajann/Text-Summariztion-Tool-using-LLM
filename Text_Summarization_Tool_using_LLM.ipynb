```python
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline("summarization", model="t5-small")
```

```python
import gradio as gr

def summarize_text(text, max_length=130, min_length=30):
    # Truncate input text to 2000 characters (optional)
    text = text[:2000]
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]['summary_text']

# Gradio interface
def gradio_summarizer(text, max_length, min_length):
    yield "Processing... Please wait."
    summary = summarize_text(text, max_length, min_length)
    yield summary

# Define inputs and outputs
inputs = [
    gr.components.Textbox(lines=10, placeholder="Enter text to summarize...", label="Input Text"),
    gr.components.Slider(minimum=50, maximum=200, value=130, label="Max Length"),
    gr.components.Slider(minimum=10, maximum=100, value=30, label="Min Length")
]
outputs = gr.components.Textbox(label="Summary")

# Create Gradio interface
interface = gr.Interface(
    fn=gradio_summarizer,
    inputs=inputs,
    outputs=outputs,
    live=True,
    title="Text Summarization Tool",
    description="Summarize long text using a pre-trained LLM."
)

# Launch the interface
interface.launch()
```
